```python
# assuming table meaning df
# assuming row 1 will always be column names:
# assuming delay times is supposed to be a array, since then flight codes increments line up 
# assuming always 2 locations specified not more
# new columns for 2. are to and from 
# assuming the first row will have a flight code always 
# assuming bottom row will not be 0 and the following values will follow the pattern so overwritting is fine
import pandas as pd 
import numpy as np
import math
```

```python
data = 'Airline Code;DelayTimes;FlightCodes;To_From\nAir Canada (!);[21, 40];20015.0;WAterLoo_NEWYork\n<Air France> (12);[];;Montreal_TORONTO\n(Porter Airways. );[60, 22, 87];20035.0;CALgary_Ottawa\n12. Air France;[78, 66];;Ottawa_VANcouvER\n""".\\.Lufthansa.\\.""";[12, 33];20055.0;london_MONTreal\n'
```

```python
dfData = []
for i in data.split('\n'):
    print(i)
    print(i.split(';'))
    tmp = i.split(';')
    dfData.append(tmp)
dfData
```

```python
table = pd.DataFrame(dfData[1:], columns=dfData[0])
table = table.fillna(0)
table
```

```python
# 1.
#table["FlightCodes"].astype(str).astype(int)  #math.floor(table["FlightCodes"].astype(int)) 
# assuming 0th index not ever 0
baseCode = int(math.floor(float(table['FlightCodes'][0])))
for index,row in table.iterrows():
    if( row["Airline Code"] != ""):
        row["FlightCodes"] = baseCode + (10*index) 
table
```

```python
# 2.
#To and From 
locations = table['To_From'].str.upper().str.split("_", n = 1, expand = True)
table = table.drop('To_From', 1)
# assuming always 2 locations specified not more
# new columns for 2. are to and from 
table["To"] = locations[0]
table["From"] = locations[1]
table
```

```python
# 3.
table['Airline Code'] = table['Airline Code'].str.replace('[^A-Za-z\s]+', '').str.strip()
table 
```
